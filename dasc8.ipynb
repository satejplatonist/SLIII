```python
import pandas as pd
import numpy as nnp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
sns.catplot(x ="sex", hue="survived",kind="count",data=df)
```

```python
group = df.groupby(['pclass','survived'])
unpack = group.size().unstack()

sns.heatmap(unpack , cbar = True , cmap = "Reds" , annot = True)
```

```python
sns.violinplot(x= "sex",y='fare',hue="survived",data =df, split=True)
```

```python
df['Fare_range'] = pd.qcut(df['fare'],4)
sns.barplot(x="Fare_range",y="survived" , data = df)
```

```python
sns.catplot(x ='embarked', hue ='survived',kind ='count', col ='pclass', data = df)
```

```python
sns.displot(df['fare'])
```

```python
sns.displot(df['fare'],kde =True)
```

```python
sns.displot(df['fare'],bins=10,kde=True)
```

```python
sns.jointplot(x="age",y="fare",data=df,kind="scatter")
```

```python
sns.jointplot(x="age",y="fare",data=df,kind="hex")
```
